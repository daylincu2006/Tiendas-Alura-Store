### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda_2 = pd.read_csv(url2)
tienda_3 = pd.read_csv(url3)
tienda_4 = pd.read_csv(url4)

tienda.head()
tienda
```

#1. Análisis de facturación



####**Calculando el ingreso por tienda**

```python
facturacion_total= 0
precios = tienda['Precio']
precios_2 = tienda_2['Precio']
precios_3 = tienda_3['Precio']
precios_4 = tienda_4['Precio']
facturacion_1 = sum(precios)
facturacion_2 = sum(precios_2)
facturacion_3 = sum(precios_3)
facturacion_4 = sum(precios_4)
facturacion_total= facturacion_1 + facturacion_2 + facturacion_3 + facturacion_4
print(f"Ingreso de Tienda 1: ${facturacion_1}")
print(f"Ingreso de Tienda 2: ${facturacion_2}")
print(f"Ingreso de Tienda 3: ${facturacion_3}")
print(f"Ingreso de Tienda 4: ${facturacion_4}")

print(f"Ingreso Total de las Tiendas: ${facturacion_total}")
```

# 2. Ventas por categoría

```python
#Analisis de ventas por categorias en las tiendas
def v_categoria(tiend):
    venta_artículos_para_el_hogar= 0
    venta_deportes_y_diversión= 0
    venta_electronico= 0
    venta_electrodomésticos= 0
    ventas_instrumentos_musicales= 0
    ventas_juguetes= 0
    ventas_libros= 0
    ventas_muebles= 0
    venta_total=0
    cantidad_artículos_para_el_hogar=0
    # Artículos para el hogar
    artículos_para_el_hogar = tiend.loc[tiend['Categoría del Producto'] == "Artículos para el hogar",'Precio']
    venta_artículos_para_el_hogar= sum(artículos_para_el_hogar)
    cantidad_artículos_para_el_hogar= artículos_para_el_hogar.count()

    #Deportes y diversión
    deportes_y_diversión = tiend.loc[tiend['Categoría del Producto'] == "Deportes y diversión",'Precio']
    venta_deportes_y_diversión= sum(deportes_y_diversión)
    cantidad_deportes_y_diversión= deportes_y_diversión.count()
    #Electronico
    electronico = tiend.loc[tiend['Categoría del Producto'] == "Electrónicos",'Precio']
    venta_electronico= sum(electronico)
    cantidad_electronico= electronico.count()
    #Electrodomésticos
    electrodomésticos = tiend.loc[tiend['Categoría del Producto'] == "Electrodomésticos",'Precio']
    venta_electrodomésticos= sum(electrodomésticos)
    cantidad_electrodomésticos= electrodomésticos.count()
    #Instrumentos musicales
    instrumentos_musicales = tiend.loc[tiend['Categoría del Producto'] == "Instrumentos musicales",'Precio']
    venta_instrumentos_musicales= sum(instrumentos_musicales)
    cantidad_instrumentos_musicales= instrumentos_musicales.count()
    #Juguetes
    juguetes = tiend.loc[tiend['Categoría del Producto'] == "Juguetes",'Precio']
    venta_juguetes= sum(juguetes)
    cantidad_juguetes= juguetes.count()
    #Libros
    libros = tiend.loc[tiend['Categoría del Producto'] == "Libros",'Precio']
    venta_libros= sum(libros)
    cantidad_libros= libros.count()
    #Muebles
    muebles = tiend.loc[tiend['Categoría del Producto'] == "Muebles",'Precio']
    venta_muebles= sum(muebles)
    cantidad_muebles= muebles.count()

    venta_total= venta_artículos_para_el_hogar + venta_deportes_y_diversión + venta_electronico + venta_electrodomésticos + ventas_instrumentos_musicales + ventas_juguetes + ventas_libros + ventas_muebles
    cantidad_total=cantidad_artículos_para_el_hogar + cantidad_deportes_y_diversión + cantidad_electronico + cantidad_electrodomésticos + cantidad_instrumentos_musicales + cantidad_juguetes + cantidad_libros + cantidad_muebles
    promedio_ventas= venta_total/cantidad_total

    datos_tiendas={'Venta':[venta_artículos_para_el_hogar,venta_deportes_y_diversión,venta_electronico,venta_electrodomésticos,venta_instrumentos_musicales,
                             venta_juguetes, venta_libros,venta_muebles, venta_total],
                   'Cantidad':[cantidad_artículos_para_el_hogar, cantidad_deportes_y_diversión, cantidad_electronico,cantidad_electrodomésticos,cantidad_instrumentos_musicales,
                               cantidad_juguetes,cantidad_libros,cantidad_muebles, cantidad_total]}

    indexes1= ['Artículos para el hogar','Deportes y diversión','Electrónicos','Electrodomésticos','Instrumentos musicales','Juguetes','Libros','Muebles','Totales']
    columns= ['Ventas','Cantidad']
    df_tienda_categorias= pd.DataFrame(data=datos_tiendas,index=indexes1)

    print(df_tienda_categorias)
    print(f"La venta promedio es {promedio_ventas:.2f}")
print(f"Venta por categoria Tienda 1")
v_categoria(tienda)
print(f"Venta por categoria Tienda 2")
v_categoria(tienda_2)
print(f"Venta por categoria Tienda 3")
v_categoria(tienda_3)
print(f"Venta por categoria Tienda 4")
v_categoria(tienda_4)
```

# 4. Productos más y menos vendidos

```python
#productos mas y menos vendidos por tienda
```

```python
def productos_con_mayores_ventas(tiend):
    inf_data = tiend.groupby(['Producto']).agg(
    prec=('Precio','sum'),
    cantidad=('Producto', 'count'))
    inf_data.sort_values('cantidad',ascending=False, inplace=True)

    print(inf_data.head(1))

def productos_con_menores_ventas(tiend):
    inf_data = tiend.groupby(['Producto']).agg(
    prec=('Precio','sum'),
    cantidad=('Producto', 'count'))
    inf_data.sort_values('cantidad',ascending=True, inplace=True)

    print(inf_data.head(1))

print(f"El producto con mayor venta en Tienda 1")
productos_con_mayores_ventas(tienda)
print(f"El producto con menor venta en Tienda 1")
productos_con_menores_ventas(tienda)

print(f"El producto con mayor venta en Tienda 2")
productos_con_mayores_ventas(tienda_2)
print(f"El producto con menor venta en Tienda 2")
productos_con_menores_ventas(tienda_2)

print(f"El producto con mayor venta en Tienda 3")
productos_con_mayores_ventas(tienda_3)
print(f"El producto con menor venta en Tienda 3")
productos_con_menores_ventas(tienda_3)

print(f"El producto con mayor venta en Tienda 4")
productos_con_mayores_ventas(tienda_4)
print(f"El producto con menor venta en Tienda 4")
productos_con_menores_ventas(tienda_4)
```
